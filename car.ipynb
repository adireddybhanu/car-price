```python
# Initialize an empty dictionary to store all the encodings
label_encodings = {}

# Loop through all object (categorical) columns in the DataFrame
for column in df.select_dtypes(include='object').columns:
    unique_values = df[column].unique()  # Get unique values for the column
    encoding = {}  # Initialize an empty dictionary for this column's encoding
    
    # Create the label encoding for this column
    for index, value in enumerate(unique_values):
        encoding[value] = index
    
    # Store the encoding in the main dictionary under the column name
    label_encodings[column] = encoding

# The label_encodings dictionary now contains all encodings for each categorical column
print(label_encodings)
```

```python
df.replace(label_encodings,inplace=True)
```

```python
# Initialize regressors
regressors = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'ElasticNet': ElasticNet(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'Support Vector Regressor': SVR(),
    'Ada Boost Regressor': AdaBoostRegressor()
}

# Train and evaluate each regressor
for name, regressor in regressors.items():
    regressor.fit(X_train, y_train)
    y_train_pred = regressor.predict(X_train)
    y_test_pred = regressor.predict(X_test)
    
    # Calculate metrics
    train_score = r2_score(y_train, y_train_pred)
    
    test_score = r2_score(y_test, y_test_pred)
    mse = mean_squared_error(y_test, y_test_pred)
    rmse = np.sqrt(mse)
    
    print(f"{name}:")
    print(f"  Training R^2 Score: {train_score:.4f}")
    print(f"  Testing R^2 Score: {test_score:.4f}")
    print(f"  Mean Squared Error: {mse:.4f}")
    print(f"  Root Mean Squared Error: {rmse:.4f}")
    print()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import randint, uniform

# Define the model
model = RandomForestRegressor()

# Define the parameter distributions for RandomizedSearchCV
param_distributions = {
    'n_estimators': randint(50, 500),                  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30, 40, 50],           # Maximum depth of the tree
    'min_samples_split': randint(2, 15),                # Minimum number of samples required to split an internal node
    'min_samples_leaf': randint(1, 8),                  # Minimum number of samples required to be at a leaf node
    'max_features': ['auto', 'sqrt', 'log2', uniform(0.1, 0.9)],  # Number of features to consider for the best split
    'bootstrap': [True, False],                        # Whether to use bootstrap samples
    'criterion': ['squared_error', 'absolute_error'],   # Criterion for splitting
    'max_samples': [None, uniform(0.5, 0.5)]           # Fraction of samples to use if bootstrap=True
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_distributions,
    n_iter=100,              # Number of parameter settings to sample
    cv=5,                   # Number of cross-validation folds
    scoring='r2',           # Scoring metric
    n_jobs=-1,              # Use all available cores
    verbose=2,              # Verbosity level
    random_state=42         # For reproducibility
)

# Fit RandomizedSearchCV
random_search.fit(X_train, y_train)

# Best parameters and best score
print("Best Hyperparameters:", random_search.best_params_)
print("Best R² Score:", random_search.best_score_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np

# Initialize lists to store results
r2_scores = []

# Perform multiple train-test splits and evaluations
for i in range(5):
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=i)
    
    # Define and train the model
    best_model = RandomForestRegressor(
        bootstrap=True,
        criterion='absolute_error',
        max_depth=50,
        max_features='log2',
        max_samples=None,
        min_samples_leaf=1,
        min_samples_split=4,
        n_estimators=150,
        random_state=42
    )
    best_model.fit(X_train, y_train)
    
    # Predict and evaluate
    y_pred = best_model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    r2_scores.append(r2)
    
    # Print results for this split
    print(f"Run {i+1} - R² Score: {r2}")

# Print mean and standard deviation of R² scores
print("Mean R² Score:", np.mean(r2_scores))
print("Standard Deviation of R² Scores:", np.std(r2_scores))
```

```python
# Save the model to a pickle file
with open('best_random_forest_model.pkl', 'wb') as file:
    pickle.dump(best_model, file)

print("Model saved to best_random_forest_model.pkl")
```
